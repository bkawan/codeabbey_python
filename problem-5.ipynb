# Minimum of Three

- Input data will contain in the first line the number of triplets to follow.
- Next lines will contain one triplet each.
- Answer should contain selected minimums of triplets, separated by spaces.

Example:

data:

3
 
7 3 5

15 20 40

300 550 137

answer:

3 15 137

# Method 1

```python
num_input = input("Enter pairs of triplet: ")

num_list = None
left_num = []
mid_num = []
right_num = []
tmp_list = []
while not num_input:
    print("Error !!Empty number!!!")
    num_input = input("Enter numbers followed by spaces to form a triplet(s) : ")

while num_input:
    try:
        num_list= [int(n) for n in num_input.split() if n]
        if len(num_list) % 3 !=0:
            print("Error in forming triplets!! total counts of number must be divisible by 3")
            num_input = input("Enter numbers followed by spaces to form a triplet(s) : ")
            continue
        break
    except ValueError:
        print("only integers  are allowed")
        num_input = input("Enter numbers followed by spaces to form a triplet(s) : ")
#     break

i = 0
while i < len(num_list):
    #print(num_list[i])

    if i%3 == 0:
        left_num.append(num_list[i])
    else:
        #tmp_list=[]
        tmp_list.append(num_list[i])
    i+=1


        #print("Temp: ",num_list[i],end=' ')
j = 0
while j< len(tmp_list):
    if j%2 ==0:
        mid_num.append(tmp_list[j])
    else:
        right_num.append(tmp_list[j])
    j+=1

k = 0
smallest = []
while k < len(num_list)/3:
    sm = left_num[k]
    if sm > mid_num[k]>right_num[k]:
        sm = right_num[k]
    elif mid_num[k]<sm :
        sm = mid_num[k]
    smallest.append(sm)
    k+=1

print("****************************************")
print("Total number of Triplets: ", len(num_list)//3)

i = 0
triplets_list=[]
print("List  of Triplets:")
while i<len(num_list):
        triplets = num_list[i:i+3]
        print()
        i +=3
        for n in triplets:
            print(n,end=" ")
print("\nSmallest numbers in triplets are: ")
for s in smallest:
    print(s,end=" ")
```
